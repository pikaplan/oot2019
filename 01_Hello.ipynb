## Simple calculation graph

We will run the calculation y = W * X as the output of the computational tensor tY using the current session object. Please note that we are multiplying an input value tX with a constant tW. 

#### Editor Settings
To show line numbers use the menu command `Tools -> Preferences -> | Editor | -> [[v]] Show line numbers`



```
import sys
import tensorflow as tf

# Info about this Collab runtime
print("[>] Python version:", sys.version)
print("[>] Tensorflow version:", tf.VERSION)

# Creation of a session using the constructor of class :Session
oSession = tf.Session()

# Creation of different classes of computational tensors using factory methods
tW = tf.constant(4.0, tf.float32)
tX = tf.placeholder(tf.float32, (None))
tY = tf.multiply(tW, tX)

# Creation of a dict using the alternative constructor of class :dict
oDict = {}
oDict[tX] = 5

# Run the calculations and print the result
nY = oSession.run(tY, feed_dict=oDict)
print("\nHello computational tensor! What is your value?", nY)

oSession.close()
```

Bellow the constant is replaced by a variable that has an initial value and there are two different sessions. In the second session we assign a different value to the variable before invoking the calculation.

```
import tensorflow as tf

nW = 4.0
tW = tf.Variable(nW, tf.float32)
tX = tf.placeholder(tf.float32, (None))
tY = tW * tX

# Create two sessions
oSession1 = tf.Session()
oSession2 = tf.Session()

# All variables need to be initialized before they can be used in a session
oSession1.run(tf.initializers.global_variables())

# Assign a different value to tW for the second session
oSession2.run(tf.initializers.global_variables())
oSession2.run(tW.assign(5.0))

# Run for the first session
nY1 = oSession1.run(tY, feed_dict={tX:5})
nY2 = oSession2.run(tY, feed_dict={tX:5})
  
print("Results, Session1: %.4f Session2: %.4f" % (nY1, nY2))
```

This shows the difference between a tensor object and the value that it holds within different sessions, in comparison the Python variable. Please note that when using a GPU the Python variable will remain stored in the main memory, while the session variables are transfered and kept on the GPU memory.

```
print("Python Variable:",nW)
print("Tensorflow Variables:",oSession1.run(tW), oSession2.run(tW))

oSession1.close()
oSession2.close()
```
